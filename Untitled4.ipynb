```python
from tensorflow.keras.utils import Sequence
from tensorflow.keras.utils import to_categorical

def create_sequences(tokenizer, max_length, captions_list, photo_feature, vocab_size):
    X1, X2, y = [], [], []
    for cap in captions_list:
        seq = tokenizer.texts_to_sequences([cap])[0]
        for i in range(1, len(seq)):
            in_seq, out_word = seq[:i], seq[i]
            in_seq = pad_sequences([in_seq], maxlen=max_length)[0]
            out_word = to_categorical([out_word], num_classes=vocab_size)[0]
            X1.append(photo_feature)
            X2.append(in_seq)
            y.append(out_word)
    return np.array(X1), np.array(X2), np.array(y)

class DataGenerator(Sequence):
    def __init__(self, image_list, captions_map, features, tokenizer, max_length, vocab_size, batch_size=32):
        self.image_list = image_list
        self.captions_map = captions_map
        self.features = features
        self.tokenizer = tokenizer
        self.max_length = max_length
        self.vocab_size = vocab_size
        self.batch_size = batch_size

    def __len__(self):
        return int(np.ceil(len(self.image_list) / float(self.batch_size)))

    def __getitem__(self, idx):
        batch_images = self.image_list[idx*self.batch_size : (idx+1)*self.batch_size]
        X1_batch, X2_batch, y_batch = [], [], []
        for img in batch_images:
            if img not in self.features or img not in self.captions_map:
                continue
            photo = self.features[img]
            caps = self.captions_map[img]
            X1c, X2c, yc = create_sequences(self.tokenizer, self.max_length, caps, photo, self.vocab_size)
            if len(X1c) > 0:
                X1_batch.extend(X1c)
                X2_batch.extend(X2c)
                y_batch.extend(yc)
        if len(X1_batch) == 0:
            # return empty batch (avoid crash)
            return [np.zeros((0,2048)), np.zeros((0,self.max_length))], np.zeros((0,self.vocab_size))
        return [np.array(X1_batch), np.array(X2_batch)], np.array(y_batch)
```

```python
def define_model(vocab_size, max_length):
    # Feature extractor
    inputs1 = Input(shape=(2048,))
    fe1 = Dropout(0.5)(inputs1)
    fe2 = Dense(256, activation='relu')(fe1)

    # Sequence model
    inputs2 = Input(shape=(max_length,))
    se1 = Embedding(vocab_size, 256, mask_zero=True)(inputs2)
    se2 = Dropout(0.5)(se1)
    se3 = LSTM(256)(se2)

    # Decoder (combine)
    decoder1 = add([fe2, se3])
    decoder2 = Dense(256, activation='relu')(decoder1)
    outputs = Dense(vocab_size, activation='softmax')(decoder2)

    model = Model(inputs=[inputs1, inputs2], outputs=outputs)
    model.compile(loss='categorical_crossentropy', optimizer='adam')
    return model

model = define_model(vocab_size, max_length)
model.summary()
```

```python
# ----------------- Fixed DataGenerator & Training Pipeline -----------------

import numpy as np
from tensorflow.keras.utils import Sequence, to_categorical
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.callbacks import ModelCheckpoint

# 1) Robust Sequence generator (returns tuple inputs, calls super, supports shuffle)
class DataGenerator(Sequence):
    def __init__(self, image_list, captions_map, features, tokenizer, max_length,
                 vocab_size, batch_size=32, shuffle=True, **kwargs):
        self.image_list = image_list
        self.captions_map = captions_map
        self.features = features
        self.tokenizer = tokenizer
        self.max_length = max_length
        self.vocab_size = vocab_size
        self.batch_size = batch_size
        self.shuffle = shuffle

        # call parent init so Keras doesn't warn
        super().__init__(**kwargs)

        self.indexes = np.arange(len(self.image_list))
        if self.shuffle:
            np.random.shuffle(self.indexes)

    def __len__(self):
        return int(np.ceil(len(self.image_list) / float(self.batch_size)))

    def on_epoch_end(self):
        if self.shuffle:
            np.random.shuffle(self.indexes)

    def __getitem__(self, idx):
        start = idx * self.batch_size
        end = (idx + 1) * self.batch_size
        batch_idx = self.indexes[start:end]
        batch_images = [self.image_list[i] for i in batch_idx]

        X1_list, X2_list, y_list = [], [], []
        for img in batch_images:
            if img not in self.features or img not in self.captions_map:
                continue
            photo = self.features[img].astype('float32')
            caps = self.captions_map[img]
            for cap in caps:
                seq = self.tokenizer.texts_to_sequences([cap])[0]
                # skip if sequence is empty for safety
                if len(seq) < 2:
                    continue
                for i in range(1, len(seq)):
                    in_seq, out_word = seq[:i], seq[i]
                    in_seq = pad_sequences([in_seq], maxlen=self.max_length)[0]
                    out_word = to_categorical([out_word], num_classes=self.vocab_size)[0]
                    X1_list.append(photo)
                    X2_list.append(in_seq)
                    y_list.append(out_word)

        # If the batch produced no training examples, return empty batch shapes
        if len(X1_list) == 0:
            return ( (np.zeros((0, 2048), dtype='float32'), np.zeros((0, self.max_length), dtype='int32')),
                     np.zeros((0, self.vocab_size), dtype='float32') )

        X1 = np.array(X1_list, dtype='float32')
        X2 = np.array(X2_list, dtype='int32')
        y  = np.array(y_list, dtype='float32')

        # Return inputs as a tuple (X1, X2) and targets y
        return ( (X1, X2), y )


# 2) Recreate model WITHOUT mask_zero to avoid LSTM/mask crashes
from tensorflow.keras.layers import Embedding, LSTM, Dense, Dropout, Input, add
from tensorflow.keras.models import Model

def define_model_safe(vocab_size, max_length):
    inputs1 = Input(shape=(2048,), name='photo_input')
    fe1 = Dropout(0.5)(inputs1)
    fe2 = Dense(256, activation='relu')(fe1)

    inputs2 = Input(shape=(max_length,), name='seq_input')
    # mask_zero=False avoids mask-related runtime asserts in some TF builds
    se1 = Embedding(input_dim=vocab_size, output_dim=256, mask_zero=False)(inputs2)
    se2 = Dropout(0.5)(se1)
    se3 = LSTM(256)(se2)

    decoder1 = add([fe2, se3])
    decoder2 = Dense(256, activation='relu')(decoder1)
    outputs = Dense(vocab_size, activation='softmax')(decoder2)

    model = Model(inputs=[inputs1, inputs2], outputs=outputs)
    model.compile(optimizer='adam', loss='categorical_crossentropy')
    return model

# Replace the old model with the safe one
model = define_model_safe(vocab_size, max_length)
model.summary()


# 3) Create generator and do a single train_on_batch sanity check before full fit
BATCH_SIZE = 32
EPOCHS = 5

generator = DataGenerator(all_images, captions, features, tokenizer, max_length, vocab_size,
                          batch_size=BATCH_SIZE, shuffle=True)

# Find first non-empty batch and run train_on_batch to catch errors early
first_ok = False
for i in range(len(generator)):
    (X1_b, X2_b), y_b = generator[i]
    if X1_b.shape[0] > 0:
        print("Sanity-check batch index:", i)
        print("X1 shape:", X1_b.shape, " dtype:", X1_b.dtype)
        print("X2 shape:", X2_b.shape, " dtype:", X2_b.dtype)
        print("y  shape:", y_b.shape, " dtype:", y_b.dtype)
        try:
            loss = model.train_on_batch([X1_b, X2_b], y_b)
            print("train_on_batch OK — loss:", loss)
            first_ok = True
        except Exception as e:
            print("train_on_batch FAILED with error:", e)
            raise
        break

if not first_ok:
    raise RuntimeError("No non-empty batch found in the generator. Check captions/features and SAMPLE_SIZE.")


# 4) Training with checkpoint
checkpoint_path = 'model-ep{epoch:02d}-loss{loss:.3f}.h5'
checkpoint = ModelCheckpoint(checkpoint_path, monitor='loss', verbose=1, save_best_only=False)

history = model.fit(generator, epochs=EPOCHS, callbacks=[checkpoint])
print("Training finished (quick test).")


# 5) Save model & tokenizer
import pickle
model.save('caption_model_safe.h5')
with open('tokenizer_safe.pkl','wb') as f:
    pickle.dump(tokenizer, f)
print("Saved caption_model_safe.h5 and tokenizer_safe.pkl")


# 6) Greedy inference helper (run to test on a random image)
from tensorflow.keras.preprocessing.sequence import pad_sequences
import random
from PIL import Image
import matplotlib.pyplot as plt

def generate_desc_greedy(model, tokenizer, photo_feature, max_length):
    in_text = 'startseq'
    for _ in range(max_length):
        seq = tokenizer.texts_to_sequences([in_text])[0]
        seq = pad_sequences([seq], maxlen=max_length)
        yhat = model.predict([photo_feature.reshape((1,2048)).astype('float32'), seq], verbose=0)
        yhat = np.argmax(yhat)
        word = tokenizer.index_word.get(yhat, None)
        if word is None:
            break
        in_text += ' ' + word
        if word == 'endseq':
            break
    return in_text

def show_image_and_caption(img_name):
    path = os.path.join(img_dir, img_name)
    img = Image.open(path)
    plt.imshow(img); plt.axis('off')
    caption = generate_desc_greedy(model, tokenizer, features[img_name], max_length)
    print("Predicted:", caption)
    print("References:")
    for c in captions.get(img_name, [])[:5]:
        print(" -", c)

# show a random test
test_img = random.choice(list(features.keys()))
show_image_and_caption(test_img)
```

```python
# ======= Safe beam-search + run on /content/images.jpg =======
import heapq
import numpy as np
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.preprocessing.image import load_img, img_to_array
from tensorflow.keras.applications.inception_v3 import preprocess_input
from PIL import Image
import matplotlib.pyplot as plt
import os

def beam_search_decoder_safe(model, tokenizer, photo_feature, max_length, beam_width=3):
    """Safer beam search: handles missing start/end tokens, dtypes, and falls back to greedy."""
    # get start/end token ids (fallbacks if not present)
    start_idx = tokenizer.word_index.get('startseq') or tokenizer.word_index.get('start') or 1
    end_idx   = tokenizer.word_index.get('endseq')   or tokenizer.word_index.get('end')   or 2

    # ensure photo_feature dtype and shape
    photo = photo_feature.astype('float32').reshape((1, -1))

    # heap items: (neg_logprob, sequence_list, completed_flag)
    heap = [(0.0, [start_idx], False)]
    completed = []

    # limit number of expansions to avoid infinite loops
    max_steps = max_length * 2

    while heap and len(completed) < beam_width and max_steps > 0:
        max_steps -= 1
        neg_logprob, seq, is_completed = heapq.heappop(heap)
        if is_completed:
            completed.append((neg_logprob, seq))
            continue

        # prepare sequence input (pad to max_length)
        seq_pad = pad_sequences([seq], maxlen=max_length, dtype='int32')
        preds = model.predict([photo, seq_pad], verbose=0)
        preds = preds.squeeze()  # shape: (vocab_size,)
        # numerical stability
        log_probs = np.log(np.clip(preds, 1e-12, 1.0))

        # pick top-k candidates
        topk = np.argsort(log_probs)[-beam_width:]
        for idx in topk:
            new_seq = seq + [int(idx)]
            new_neg = neg_logprob - float(log_probs[idx])
            if int(idx) == int(end_idx):
                completed.append((new_neg, new_seq))
            else:
                heapq.heappush(heap, (new_neg, new_seq, False))

        # keep heap bounded
        if len(heap) > beam_width * 10:
            heap = heapq.nsmallest(beam_width * 10, heap)

    # choose best completed seq
    if completed:
        completed.sort(key=lambda x: x[0])
        best = completed[0][1]
    else:
        # fallback: greedy decode
        best = [start_idx]
        for _ in range(max_length):
            seq_pad = pad_sequences([best], maxlen=max_length, dtype='int32')
            preds = model.predict([photo, seq_pad], verbose=0).squeeze()
            nxt = int(np.argmax(preds))
            best.append(nxt)
            if nxt == end_idx:
                break

    # convert indices to words (skip start/end tokens)
    words = []
    for ind in best:
        w = tokenizer.index_word.get(ind)
        if w is None:
            continue
        if w in ('startseq','endseq','start','end'):
            continue
        words.append(w)
    return ' '.join(words).strip()

# ---- extract features for uploaded image /content/images.jpg ----
img_path = '/content/images.jpg'
if not os.path.exists(img_path):
    raise FileNotFoundError(f"{img_path} not found. Upload your image first or adjust path.")

def extract_feature_for_inference(path, model):
    img = load_img(path, target_size=(299,299))
    x = img_to_array(img)
    x = np.expand_dims(x, axis=0)
    x = preprocess_input(x)
    feat = model.predict(x, verbose=0)
    return feat[0]

# use your existing feature_extractor variable (InceptionV3 model -> pooled features)
photo_feature = extract_feature_for_inference(img_path, feature_extractor)
print("Feature shape:", photo_feature.shape, " dtype:", photo_feature.dtype)

# ---- run beam search (safe) ----
caption = beam_search_decoder_safe(model, tokenizer, photo_feature, max_length, beam_width=3)
print("\nPredicted caption:", caption)

# ---- display image ----
img = Image.open(img_path)
plt.imshow(img); plt.axis('off')

```

```python
  from tensorflow.keras.utils import Sequence, to_categorical
from tensorflow.keras.preprocessing.sequence import pad_sequences
import numpy as np

class DataGenerator(Sequence):
    def __init__(self, image_list, captions_map, features, tokenizer, max_length,
                 vocab_size, batch_size=32, shuffle=True, **kwargs):

        self.image_list = image_list
        self.captions_map = captions_map
        self.features = features
        self.tokenizer = tokenizer
        self.max_length = max_length
        self.vocab_size = vocab_size
        self.batch_size = batch_size
        self.shuffle = shuffle

        super().__init__(**kwargs)

        self.indexes = np.arange(len(self.image_list))
        if self.shuffle:
            np.random.shuffle(self.indexes)

    def __len__(self):
        return int(np.ceil(len(self.image_list) / self.batch_size))

    def on_epoch_end(self):
        if self.shuffle:
            np.random.shuffle(self.indexes)

    def __getitem__(self, idx):
        start = idx * self.batch_size
        end = (idx + 1) * self.batch_size

        batch_idx = self.indexes[start:end]
        batch_imgs = [self.image_list[i] for i in batch_idx]

        X1, X2, Y = [], [], []

        for img in batch_imgs:
            if img not in self.features or img not in self.captions_map:
                continue

            photo = self.features[img].astype('float32')
            caps = self.captions_map[img]

            for cap in caps:
                seq = self.tokenizer.texts_to_sequences([cap])[0]
                if len(seq) < 2:
                    continue

                for i in range(1, len(seq)):
                    in_seq = seq[:i]
                    out_word = seq[i]

                    in_seq = pad_sequences([in_seq], maxlen=self.max_length)[0]
                    out_word = to_categorical([out_word], num_classes=self.vocab_size)[0]

                    X1.append(photo)
                    X2.append(in_seq)
                    Y.append(out_word)

        if len(X1) == 0:
            return ((np.zeros((0,2048)), np.zeros((0,self.max_length))), np.zeros((0,self.vocab_size)))

        return ((np.array(X1), np.array(X2)), np.array(Y))
```

```python
# =========================
# UPLOAD IMAGE → GET CAPTION
# =========================

from google.colab import files
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
import os
from tensorflow.keras.preprocessing.image import load_img, img_to_array
from tensorflow.keras.applications.inception_v3 import preprocess_input
from tensorflow.keras.preprocessing.sequence import pad_sequences
import heapq

# ---- 1. Extract features using your trained InceptionV3 encoder ----
def extract_feature(path, feature_extractor):
    img = load_img(path, target_size=(299, 299))
    x = img_to_array(img)
    x = np.expand_dims(x, axis=0)
    x = preprocess_input(x)
    feature = feature_extractor.predict(x, verbose=0)
    return feature[0]

# ---- 2. Beam search decoder (simple + clean) ----
def beam_search(model, tokenizer, photo_feature, max_length, beam_width=3):
    start = tokenizer.word_index['startseq']
    end = tokenizer.word_index['endseq']

    sequences = [[ [start], 0.0 ]]  # [sequence, score]

    for _ in range(max_length):
        temp = []
        for seq, score in sequences:
            seq_pad = pad_sequences([seq], maxlen=max_length)
            yhat = model.predict([photo_feature.reshape(1,2048), seq_pad], verbose=0)[0]

            # take top-k words
            top_k = np.argsort(yhat)[-beam_width:]

            for word in top_k:
                new_seq = seq + [word]
                new_score = score + np.log(yhat[word] + 1e-7)

                temp.append([new_seq, new_score])

        # sort by score
        sequences = sorted(temp, key=lambda x: x[1], reverse=True)
        sequences = sequences[:beam_width]  # keep best beams

    # best sequence
    best_seq = sequences[0][0]

    # convert tokens → words
    caption = []
    for word in best_seq:
        mapped = tokenizer.index_word.get(word)
        if mapped == 'endseq':
            break
        if mapped not in ['startseq', 'endseq', None]:
            caption.append(mapped)
    return ' '.join(caption)

# ---- 3. Upload image + run captioning ----
def caption_image():
    uploaded = files.upload()              # upload
    fname = list(uploaded.keys())[0]       # get filename

    # extract features
    photo_feature = extract_feature(fname, feature_extractor)

    # generate caption
    caption = beam_search(model, tokenizer, photo_feature, max_length)

    # display
    img = Image.open(fname)
    plt.imshow(img)
    plt.axis("off")
    print("\n📌 Caption:", caption)

    return caption

print("Ready! Run caption_image() to upload an image and get caption.")
```
